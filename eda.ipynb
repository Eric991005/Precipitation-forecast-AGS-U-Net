```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib import cm  # 导入cm模块
import os
import random

# 设定主文件夹路径
main_folder_path = '/root/autodl-tmp/competition/data/NJU_CPOL_kdpRain'

# 获取所有子目录的列表
all_subdirs = [d for d in os.listdir(main_folder_path) if os.path.isdir(os.path.join(main_folder_path, d))]

# 随机选择5个子目录
random_selected_subdirs = random.sample(all_subdirs, 5)

# 遍历每个选定的子目录
for selected_subdir in random_selected_subdirs:
    # 初始化一个空数组来存储降雨量总和
    rainfall_sum = np.zeros((256, 256))

    # 设置当前子目录的路径
    folder_path = os.path.join(main_folder_path, selected_subdir)

    # 遍历当前子目录下的所有.npy文件
    for file_name in sorted(os.listdir(folder_path)):
        if file_name.endswith('.npy'):
            # 加载每个.npy文件中的数据
            file_path = os.path.join(folder_path, file_name)
            data = np.load(file_path)
            # 将每个文件中的数据加到rainfall_sum数组中
            rainfall_sum += data

    # 创建一个自定义的颜色映射
    colors = [(1, 1, 1)] + [(cm.jet(i)) for i in range(1, 256)]
    new_map = mcolors.LinearSegmentedColormap.from_list('new_map', colors, N=256)

    # 用imshow显示降雨量总和
    plt.figure(figsize=(10, 10))
    plt.imshow(rainfall_sum, cmap=new_map, origin='upper')
    plt.colorbar(label='Total Rainfall (mm/hr)')
    plt.title(f'Total Rainfall Map for Rain Event in {selected_subdir}')
    plt.show()
```

```python
## 所有场雨
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib import cm  # Import cm module
import os

# 初始化一个空数组来存储降雨量总和
rainfall_sum = np.zeros((256, 256))

# 设定主文件夹路径
main_folder_path = '/root/autodl-tmp/competition/data/NJU_CPOL_kdpRain'

# 遍历所有的data_dir_XXX目录
for dir_name in sorted(os.listdir(main_folder_path)):
    dir_path = os.path.join(main_folder_path, dir_name)
    if os.path.isdir(dir_path):
        # 遍历每个目录下的所有.npy文件
        for file_name in sorted(os.listdir(dir_path)):
            if file_name.endswith('.npy'):
                # 加载每个.npy文件中的数据
                file_path = os.path.join(dir_path, file_name)
                data = np.load(file_path)
                # 将每个文件中的数据加到rainfall_sum数组中
                rainfall_sum += data

# 创建一个自定义的颜色映射
colors = [(1, 1, 1)] + [(cm.jet(i)) for i in range(1, 256)]
new_map = mcolors.LinearSegmentedColormap.from_list('new_map', colors, N=256)

# 用imshow显示降雨量总和
plt.figure(figsize=(10, 10))  # 设定图像大小
plt.imshow(rainfall_sum, cmap=new_map, origin='upper')
plt.colorbar(label='Total Rainfall (mm/hr)')
plt.title('Total Rainfall Map for All Rain Events')
plt.show()
```

```python
# 计算整个256x256图像的像素值的平均值和标准差
mean_value = np.mean(rainfall_sum)
std_value = np.std(rainfall_sum)

# 提取中心区域的16x16像素
center_x, center_y = 128, 128  # 256x256图像的中心点
half_size = 12  # 16x16区域的一半大小
x_start, x_end = center_x - half_size, center_x + half_size
y_start, y_end = center_y - half_size, center_y + half_size
center_area = rainfall_sum[x_start:x_end, y_start:y_end]

# 定义一个阈值来识别异常点
threshold_upper = mean_value + 3 * std_value
threshold_lower = mean_value - 3 * std_value

# 找到所有异常点的坐标
abnormal_points = np.argwhere((center_area > threshold_upper) | (center_area < threshold_lower))

if len(abnormal_points) > 0:
    # 根据异常点的坐标，计算出包含这些点的最小矩形的边界
    min_x = np.min(abnormal_points[:, 0])
    max_x = np.max(abnormal_points[:, 0])
    min_y = np.min(abnormal_points[:, 1])
    max_y = np.max(abnormal_points[:, 1])

    # 根据边界计算出矩形的边长
    rect_width = max_x - min_x + 1
    rect_height = max_y - min_y + 1

    print(f"The rectangle containing all abnormal points has width: {rect_width}, height: {rect_height}")
else:
    print("No abnormal points found in the center 16x16 area.")
```

```python
import numpy as np
import os

# 定义中心区域的边界
center_x, center_y = 128, 128  # 256x256图像的中心点
radius = 9  # 圆的半径

# 设定主文件夹路径
main_folder_path = '/root/autodl-tmp/competition/data/NJU_CPOL_update2308/dBZ/3.0km'

# 创建新的文件夹来存放处理好的数据
processed_folder_path = '/root/autodl-tmp/competition/data/NJU_CPOL_update2308/dBZ/Processed_3.0km'
os.makedirs(processed_folder_path, exist_ok=True)

# 创建一个坐标网格来检查每个像素是否在圆内
y, x = np.ogrid[-radius: radius+1, -radius: radius+1]
mask = x**2 + y**2 <= radius**2

# 遍历所有的data_dir_XXX目录
for dir_name in sorted(os.listdir(main_folder_path)):
    dir_path = os.path.join(main_folder_path, dir_name)
    if os.path.isdir(dir_path):
        # 在新的文件夹中创建相应的子文件夹
        new_dir_path = os.path.join(processed_folder_path, dir_name)
        os.makedirs(new_dir_path, exist_ok=True)
        # 遍历每个目录下的所有.npy文件
        for file_name in sorted(os.listdir(dir_path)):
            if file_name.endswith('.npy'):
                # 加载每个.npy文件中的数据
                file_path = os.path.join(dir_path, file_name)
                data = np.load(file_path)
                # 获取中心区域的数据
                center_area = data[center_x-radius:center_x+radius+1, center_y-radius:center_y+radius+1][mask]
                # 检查是否70%以上的像素值为0
                zero_pixels_ratio = np.sum(center_area == 0) / center_area.size
                if zero_pixels_ratio > 0.8:
                    # 将这些像素点都设置为0
                    data[center_x-radius:center_x+radius+1, center_y-radius:center_y+radius+1][mask] = 0
                # 保存处理过的数据到新的文件夹
                new_file_path = os.path.join(new_dir_path, file_name)
                np.save(new_file_path, data)

print('Processing completed.')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import random

# 初始化变量
max_rainfall = float('-inf')  # 初始化为负无穷，确保任何实际值都会更新它
min_rainfall = float('inf')   # 初始化为正无穷，确保任何实际值都会更新它
all_rainfall_values = []  # 用于存储所有降雨量值的列表

# 设定主文件夹路径
main_folder_path = '/root/autodl-tmp/competition/data/Processed_NJU_CPOL_kdpRain'

# 获取所有的data_dir_XXX目录
all_dirs = [dir_name for dir_name in sorted(os.listdir(main_folder_path)) if os.path.isdir(os.path.join(main_folder_path, dir_name))]

# 随机选取10个目录
selected_dirs = random.sample(all_dirs, 1)

# 遍历选取的data_dir_XXX目录
for dir_name in selected_dirs:
    dir_path = os.path.join(main_folder_path, dir_name)
    # 遍历每个目录下的所有.npy文件
    for file_name in sorted(os.listdir(dir_path)):
        if file_name.endswith('.npy'):
            # 加载每个.npy文件中的数据
            file_path = os.path.join(dir_path, file_name)
            data = np.load(file_path)
            # 更新最大和最小降雨量
            max_rainfall = max(max_rainfall, np.max(data))
            min_rainfall = min(min_rainfall, np.min(data))
            # 收集所有降雨量值
            all_rainfall_values.extend(data.flatten())

# 输出最大和最小降雨量
print(f'Max Rainfall: {max_rainfall} mm/hr')
print(f'Min Rainfall: {min_rainfall} mm/hr')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import random

# 设定雨量分类
bins_ranges = [(0, 0.5), (0.5, 2), (2, 5), (5, 10), (10, 30), (30, float('inf'))]
bins_labels = ['No/Hardly noticeable', 'Light', 'Light to moderate', 'Moderate', 'Moderate to heavy', 'Rainstorm warning']

# 初始化一个数组来存储每个级别的频率
rainfall_levels_count = np.zeros(len(bins_ranges))

# 设定主文件夹路径
main_folder_path = '/root/autodl-tmp/competition/data/Processed_NJU_CPOL_kdpRain'

# 获取所有的data_dir_XXX目录
all_dirs = [dir_name for dir_name in sorted(os.listdir(main_folder_path)) if os.path.isdir(os.path.join(main_folder_path, dir_name))]

# 随机选择10个目录
selected_dirs = random.sample(all_dirs, 10)

# 遍历选定的data_dir_XXX目录
for dir_name in selected_dirs:
    dir_path = os.path.join(main_folder_path, dir_name)
    # 遍历每个目录下的所有.npy文件
    for file_name in sorted(os.listdir(dir_path)):
        if file_name.endswith('.npy'):
            # 从每个.npy文件加载数据
            file_path = os.path.join(dir_path, file_name)
            data = np.load(file_path)
            # 将每个雨量值分类
            for value in data.flatten():
                for i, (low, high) in enumerate(bins_ranges):
                    if low <= value < high:
                        rainfall_levels_count[i] += 1
                        break

# 计算总数以及每个级别的比率
total_count = np.sum(rainfall_levels_count)
rainfall_levels_ratio = (rainfall_levels_count / total_count) * 100

# 绘制每个雨量级别的柱状图
plt.figure(figsize=(10, 6))
bars = plt.bar(bins_labels, rainfall_levels_count, color=['#d9d9d9', '#a6cee3', '#1f78b4', '#b2df8a', '#33a02c', '#fb9a99'])

# 在每个柱上添加比率标签
for bar, ratio in zip(bars, rainfall_levels_ratio):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{ratio:.2f}%', ha='center', va='bottom', fontsize=10)

plt.title('Rainfall Levels Frequency')
plt.xlabel('Rainfall Level')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')  # Rotate x-labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjusts subplot params to give specified padding
plt.show()
```
